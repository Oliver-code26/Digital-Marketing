# Section 2

Section 2 content

```python
import matplotlib.pyplot as plt

for campaign in campaign_summary.index:
    values = campaign_summary.loc[campaign].values
    labels = campaign_summary.columns
    sizes = values / values.sum() * 100

    plt.figure(figsize=(6, 6))
    plt.pie(
        sizes,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90,
        wedgeprops={'edgecolor': 'white'}
    )
    plt.title(f'Pie Chart for {campaign}')
    plt.axis('equal')
    plt.show()
```
